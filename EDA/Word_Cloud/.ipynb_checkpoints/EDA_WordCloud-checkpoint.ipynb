```python
import pandas as pd
import matplotlib.pyplot as plt
import wordcloud
```

```python
df = pd.read_csv('아편_tokenized_final.csv', encoding = 'utf-8')
```

```python
print(df.columns)
```

```python
#Okt

from konlpy.tag import Okt
morphs = []
okt = Okt()

for tweet in df['content']:
    morphs.append(okt.pos(tweet))
    
word_list = []
for i in morphs:
    for word, tag in i:
        if tag in['Noun']:  #명사
            word_list.append(word)

import collections

count = collections.Counter(word_list)

words = (dict(count.most_common()))
print(words)
```

```python
#Komoran

from konlpy.tag import Komoran
morphs = []
komoran = Komoran()

for tweet in df['content']:
    morphs.append(komoran.pos(tweet))
    
word_list = []
for i in morphs:
    for word, tag in i:
        if tag in['NNG', 'NNP']:  #보통명사, 고유명사
            word_list.append(word)

import collections

count = collections.Counter(word_list)

words = (dict(count.most_common()))
print(words)
```

```python
#Hannanum: 너무 자잘하게 나옴

from konlpy.tag import Hannanum
morphs = []
hannanum = Hannanum()

for tweet in df['content']:
    morphs.append(hannanum.pos(tweet))
    
word_list = []
for i in morphs:
    for word, tag in i:
        if tag in['N']:  #명사
            word_list.append(word)

import collections

count = collections.Counter(word_list)

words = (dict(count.most_common()))
print(words)
```

```python
#Kkma: 제일 느림

from konlpy.tag import Kkma
morphs = []
kkma = Kkma()

for tweet in df['content']:
    morphs.append(kkma.pos(tweet))
    
word_list = []
for i in morphs:
    for word, tag in i:
        if tag in['NNG', 'NNP']:  #보통명사, 고유명사
            word_list.append(word)

import collections

count = collections.Counter(word_list)

words = (dict(count.most_common()))
print(words)
```

```python
#Mecab: 짱 빠름

from konlpy.tag import Mecab
morphs = []
mecab = Mecab()

for tweet in df['content']:
    morphs.append(mecab.pos(tweet))
    
word_list = []
for i in morphs:
    for word, tag in i:
        if tag in['NNG', 'NNP']:  #보통명사, 고유명사
            word_list.append(word)

import collections

count = collections.Counter(word_list)

words = (dict(count.most_common()))
print(words)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

#import nltk
#from nltk.corpus import stopwords

#okt, komoran, hannanum, kkma, mecab
palettes = ['summer', 'seismic','PuBu', 'cool', 'autumn']

wordcloud = WordCloud(font_path = 'GmarketSansTTFMedium.ttf', background_color = 'white', colormap = palettes[4],
                     width = 3000, height = 2000).generate_from_frequencies(words)

plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```

```python
wordcloud.to_file('아편_mecab.png')
```
