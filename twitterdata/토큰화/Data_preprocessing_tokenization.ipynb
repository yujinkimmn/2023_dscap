```python
from konlpy.tag import Okt, Komoran, Hannanum, Kkma

def tokenize(data):
    okt = Okt()
    komoran = Komoran()
    hannanum = Hannanum()
    kkma = Kkma()
    
    # tokenizer ì¢…ë¥˜
    tokenizers = [okt, komoran, hannanum, kkma]
    
    # í† í°í™” ê²°ê³¼ë¥¼ ì €ì¥í•  dictionary
    result = {}
    
    # ê°ê°ì˜ tokenizer ì´ìš©í•´ì„œ í† í°í™” í›„ resultì— ì €ì¥
    for tokenizer in tokenizers:
        name = str(tokenizer)
        result[name] = data['content'].apply(lambda x: tokenizer.pos(x))
    
    # ì˜¤ë¦¬ì§€ë„ ë°ì´í„°ì™€ í† í°í™” ê²°ê³¼ë¥¼ í•©ì¹¨
    return pd.concat([data] + list(result.values()), axis=1)
```
