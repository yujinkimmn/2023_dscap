## 마약 거래 필터링 파일과 원본 파일 대조하여 마약 거래 게시물의 인덱스 값 찾기

```python
import os
path = '/Users/stella/github/2023_dscap/twitterdata/original(필터링 전)'
dir = os.listdir(path)
dir
```

```python
dir = [
 '메스암페타민',
 '사일로시빈',
 '대마',
 'MDMA',
 'LSD',
 'GHB',
 '코카인',
 '아편']

dir
```

```python
케타민 = pd.read_csv('/Users/stella/github/2023_dscap/twitterdata/original(필터링 전)/케타민/케타민.csv', lineterminator='\n')
```

```python
케타민
```

```python
# 모든 CSV 파일을 담을 빈 데이터프레임 생성
all_data = pd.DataFrame()
# 각 디렉토리에 접근하여 CSV 파일 불러오기
for d in dir:
    #print(d)
    # 디렉토리 내 CSV 파일 목록 가져오기
    files = [f for f in os.listdir(os.path.join(path, d)) if f.endswith(".csv")]
    #print(files)
    
    # 각 CSV 파일을 데이터프레임으로 읽어와서 결합
    for file in files:
        file_path = os.path.join(path, d, file)
        print(file_path)
        df = pd.read_csv(file_path, encoding='utf-8')
        all_data = pd.concat([all_data, df])


# 결과 출력
print(len(all_data))
```

```python
all_data
```

```python
all_data = pd.concat([all_data, 캔디케이])
all_data = pd.concat([all_data, 케타민])
all_data
```
